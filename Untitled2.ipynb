```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flip = pd.read_csv('flipkart_com-ecommerce_sample.csv')
flip.head(5)
```

```python
flip.info()
```

```python
ama = pd.read_csv("amz_com-ecommerce_sample.csv", encoding="latin1")
ama.head(5)
```

```python
ama.info()
```

```python
#datacompy takes two dataframes as input and gives us a human-readable report containing statistics that lets us know the 
#similarities and dissimilarities between the two dataframes.

import datacompy

compare = datacompy.Compare(flip, ama, join_columns='product_name', abs_tol=0.0001, rel_tol=0)
print(compare.report())
```

```python
flip.drop(flip.iloc[:, 8:14], inplace=True, axis=1)
flip.head(5)
```

```python
flip.drop(['uniq_id','crawl_timestamp','product_url','product_category_tree','pid'], axis=1, inplace = True)
flip
```

```python
ama.drop(ama.iloc[:, 8:14], inplace=True, axis=1)
ama.head(5)
```

```python
ama.drop(['uniq_id','crawl_timestamp','product_url','product_category_tree','pid','product_specifications'], axis=1, inplace = True)
ama
```
